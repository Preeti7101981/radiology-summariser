```python
# data_processing.py

import os
import pandas as pd
from lxml import etree

records = []

folder = "data/raw"

for file in os.listdir(folder):
    if file.endswith(".xml"):

        tree = etree.parse(os.path.join(folder,file))

        findings = ""
        impression = ""

        for node in tree.xpath("//AbstractText"):

            label = node.get("Label")

            if label == "FINDINGS":
                findings = node.text

            elif label == "IMPRESSION":
                impression = node.text

        if findings and impression:
            records.append({
                "findings": findings,
                "impression": impression
            })

df = pd.DataFrame(records)

df.to_csv(
    "data/processed/radiology_reports.csv",
    index=False
)

print(df.shape)
```
